# Ways to Create Set 


```python
my_set = set()
# Returns the type of Set
type(my_set)
my_set_1 = set([""])
print(my_set)
print(my_set_1)
```

## Set can contain Multiple Data Types integer, string etc.,

```python
my_set = {1,2,3,4,5,'123', 'Data' ,'Science','Kaggle'}
print (my_set)
```

#Add Elements to a set 

```python
my_set.add('Analytics Vidhya')
print(my_set)
#Find the size of the Set 
len(my_set)
```

## Update the elements in a set 

```python
print('Before the Run : ',my_set)
my_set.update()
print('After the Run  : ',my_set)
my_set_2 = {'Award','Winner','Data', 'Kaggle', 'Science'}
print ('My Second Set :: ',my_set_2)
print('Before the Run my_set: ',my_set)
print('Before the Run my_set_2: ',my_set_2)
print('Before the Run my_set Size: ',len(my_set))
print('Before the Run my_set_2 Size: ',len(my_set_2))
my_set.update(my_set_2)
print ('my_set :: ',my_set)
print('After the Run my_set Size: ',len(my_set))
print('After the Run my_set_2 Size: ',len(my_set_2))
```

##Special Case of update function in Set (If any passed as string, each character is treated as separate set values)

```python
my_set.clear()
print('Before the Run : ',my_set)
my_set.update('This is Rolex')
print('After the Run  : ',my_set)
```

## Remove the element in the Set using pop  

```python
my_set.pop() #
print(my_set)
```

## Remove the element in the Set using remove  

```python
print('Before the Run : ',my_set)
my_set.remove('123') 
print('After the Run : ',my_set)
#my_set.remove('Not available element') # Expected use to throw error <<If you don't know that the element is availabe or not please use discard>>
```

## Exception handling for the above error in Python


```python
try:
  print('Before the Run : ',my_set)
  my_set.remove('123') 
  print('After the Run : ',my_set)
except (TypeError, ZeroDivisionError,KeyError):
   # Handle multiple exceptions
   # TypeError, ZeroDivisionError and KeyError
   print('Element you tried to remove doen\'t available in the Set')
   pass
```

##  Remove the element in the Set using discard

```python
print('Before the Run : ',my_set)
my_set.discard('Science') 
print('After the Run : ',my_set)
print('Before the Run : ',my_set)
my_set.discard('Not available element')
print('After the Run : ',my_set)
```

## Remove the element in the Set using clear

```python
print('Before the Run : ',my_set)
my_set.clear() 
print('After the Run : ',my_set)
```
