In this Note Book we plan to chcek the Date and Time Funcitons in Pandas

```python
from datetime import datetime
import pandas as pd
def date_attributes (my_date):
  print(type(my_date))
  print('Default Format ',my_date)
  print('Day ',my_date.day)
  print('Month ',my_date.month)
  print('year ',my_date.year)
  print('Format Date/Month/Year ',my_date.strftime("%d/%m/%Y"))
  print('Format Month/Date/Year ',my_date.strftime("%m/%d/%Y"))
  print('Format Month(Texual)/Date/Year ',my_date.strftime("%B/%d/%Y"))
  print('Format Year/Month/Date ',my_date.strftime("%Y/%m/%d"))
  return
my_date = datetime(year=2015, month=7, day=4)
date_attributes(my_date)
date_attributes(datetime.now())
sys_date_time=pd.to_datetime(datetime.now())
date_attributes(sys_date_time)
print (type(sys_date_time))
print (sys_date_time)
```

Let's Load Raw Data and try out various Date and Timestamp related Functions

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
df_sales=pd.read_csv('../input/retaildataset/sales data-set.csv')
df_sales
```

```python
df_sales.describe()
df_sales.dtypes
```

Here in the dataframe we see the Date column is marked as an object hence date function cannot be applied on this column, we will need to convert to date datatype to effectivley use the data time functions in pandas

```python
df_sales['Date']= pd.to_datetime(df_sales['Date'])
df_sales.dtypes
```

```python
df_sales['Year of Sales']= df_sales['Date'].dt.year
df_sales['Month of Sales']= df_sales['Date'].dt.month
df_sales['Day of Sales']= df_sales['Date'].dt.day
df_sales[['Date','Day of Sales', 'Month of Sales','Year of Sales']].head(10)
```

Below is to create a data range using Pandas, here we try to create a calendar for 2023

```python
dt_index_year_2023 = pd.date_range(start='1/1/2023', end='31/12/2023')
type(dt_index_year_2023)
print (dt_index_year_2023)
```

```python
type(dt_index_year_2023)
```

```python
idx = pd.Index(dt_index_year_2023, name='Day of the Year')
df_2023 = idx.to_frame()
type(df_2023)
df_2023.describe(datetime_is_numeric=True)
df_2023
```

```python
df_2023['DAY']=df_2023['Day of the Year'].dt.day_name()
df_2023['Month']=df_2023['Day of the Year'].dt.month_name()
df_2023
```
